# 1. CASO DE NEGOCIO 

## 1.1. Descripción del Problema 
En el entorno actual del comercio minorista (Retail), la gestión ineficiente de la cadena de suministro y la falta de visibilidad en tiempo real de los inventarios representan desafíos críticos que impactan directamente la rentabilidad y la satisfacción del cliente. Actualmente, la organización enfrenta dificultades para consolidar y procesar de manera ágil grandes volúmenes de datos distribuidos en múltiples fuentes de información dispersas: catálogos de productos (`ITEMS`), datos maestros de tiendas (`SITES`), registros históricos de transacciones (`BASE`) y niveles actuales de existencias (`STOCK`).

Esta desconexión y la falta de analítica avanzada provocan dos problemas operacionales severos:
1. **Quiebres de Stock (Stockouts):** Desabastecimiento de productos de alta demanda sin una priorización clara de reposición, lo que se traduce en ventas perdidas y migración de clientes hacia la competencia.
2. **Exceso de Inventario (Overstock):** Acumulación ineficiente de capital de trabajo y costos de almacenamiento en productos de baja rotación, debido a la ausencia de umbrales máximos y mínimos dinámicos por cada binomio artículo-tienda (Item-Site).

La falta de un sistema predictivo y centralizado impide reaccionar proactivamente ante los patrones de venta de los consumidores y evaluar el desempeño de los proveedores de manera automatizada.

## 1.2. Objetivos del Proyecto 
El objetivo principal de este proyecto es implementar una solución integral de Big Data en **Databricks**, utilizando una **Arquitectura Medallion** (Bronze, Silver y Gold) para centralizar, transformar y explotar los datos de inventario y ventas, transformándolos en decisiones estratégicas de compra.

[cite_start]Con el análisis de datos masivos y el despliegue de un pipeline automatizado, se espera lograr el siguiente impacto[cite: 6]:
* **Optimización de Disponibilidad:** Identificar y priorizar de forma automática los *stockouts* de productos con alta demanda para mitigar el impacto financiero en el negocio.
* **Abastecimiento Proactivo e Inteligente:** Desarrollar un modelo analítico capaz de proyectar alertas de compra anticipadas basándose en los patrones históricos de venta (`BASE`), evitando llegar al quiebre de stock.
* **Control Eficiente de Inventarios:** Clasificar dinámicamente cada relación *Item-Site* mediante políticas de inventario ideal (Mínimos y Máximos) para detectar excesos de stock y necesidades críticas de reabastecimiento.
* **Desarrollo de un Dashboard de Control Centralizado:** Proveer un tablero de control interactivo para los tomadores de decisiones que unifique las alertas operativas y un *Ranking de Proveedores* automatizado para negociaciones más estratégicas.
* **Reducción de Costos Operativos:** Minimizar las horas-hombre dedicadas a la planeación manual de compras y optimizar los costos logísticos derivados del almacenamiento de excesos.

## 3.2. Explicación Técnica de los Componentes de Nube
Para dar soporte a la solución de optimización de compras y gestión de inventarios, se propone una arquitectura de **Lakehouse** unificada en la nube. Aunque el diseño empresarial ideal contempla una conexión directa y automatizada, el entorno actual ha sido adaptado mediante una zona de aterrizaje temporal debido a restricciones de acceso de red y políticas de seguridad corporativas:

1. **Origen de Datos Transaccional (SQL Server):** Representa el sistema de origen central (ERP o WMS de la compañía) donde residen originalmente las tablas operativas del negocio. Debido a las restricciones técnicas actuales, los datos son extraídos de forma asíncrona hacia archivos planos con formato `.csv`.
2. **Capa de Almacenamiento (Cloud Object Storage - Landing Zone):** Repositorio de objetos en la nube que actúa como la zona de aterrizaje para las extracciones de SQL Server (`ITEMS.csv`, `SITES.csv`, `BASE.csv` y `STOCK.csv`) antes de que el motor analítico inicie su lectura.
3. **Entorno de Cómputo Analítico (Databricks Workspace):** Entorno unificado basado en **Apache Spark** que se encarga de realizar la lectura distribuida de los archivos, la limpieza pesada, los cruzados relacionales (`JOINs`) y el procesamiento de las reglas de negocio a gran escala.
4. **Formato de Almacenamiento Lógico (Delta Lake):** Tecnología de almacenamiento de código abierto que corre sobre el Object Storage e introduce características de bases de datos relacionales tradicionales, tales como transacciones ACID, control de versiones (Time Travel) y validación de esquemas.
5. **Capa de Visualización y Consumo (Databricks SQL & Dashboards):** Herramienta nativa integrada que consumirá directamente las tablas agregadas de la capa Gold para renderizar de manera interactiva las métricas del negocio y disparar las alertas operativas.

## 3.3. Flujo de Información (Estrategia Medallion)
El refinamiento y la maduración de los datos procedentes de SQL Server se estructuran en tres niveles lógicos dentro de Databricks:

* **Fase de Ingesta (Capa Bronze - Datos Crudos):** Los archivos CSV se importan respetando estrictamente sus formatos de origen (manejando los delimitadores por coma y punto y coma correspondientes) y se persisten como tablas Delta nativas. No se altera la información para mantener el linaje completo y permitir auditorías ante fallos.
* **Fase de Enriquecimiento (Capa Silver - Limpieza y Cruces):** En esta etapa se ejecuta la tipificación estricta de las columnas. Se corrigen las anomalías regionales de formato (conversión de texto con comas decimales a números numéricos de tipo `Double` o `Float`) y se resuelven las discrepancias de consistencia lógica entre el histórico de ventas (`BASE`), las existencias de inventario (`STOCK`), y los catálogos maestros (`ITEMS` y `SITES`).
* **Fase de Negocio (Capa Gold - Agregaciones y Alertas):** Las tablas enriquecidas se transforman en indicadores directamente accionables por el área de compras. Se calculan los niveles críticos por tienda-producto (Stockouts y Excesos), los umbrales de inventario ideal (mínimos y máximos) y se consolida el *Ranking de Proveedores*.

## 3.4. Especificaciones del Clúster de Cómputo (Databricks Compute)
Para garantizar que el procesamiento de las transacciones (especialmente los miles de registros del histórico en `BASE`) se realice de manera eficiente, se define la siguiente configuración de infraestructura:

* **Política de Terminación Automática:** Configurada a 20 minutos de inactividad para optimizar el consumo de créditos de nube cuando el notebook o tablero no esté en uso.

# 4. PIPELINE DE INGESTA DE DATOS 

En esta sección se desarrolla el núcleo técnico del proyecto utilizando una **Arquitectura Medallion** (Bronze, Silver y Gold). El objetivo es automatizar y estructurar de manera óptima el procesamiento de las cuatro fuentes de información extraídas desde el entorno de SQL Server (`ITEMS`, `SITES`, `BASE` y `STOCK`), las cuales se localizan unificadas en la carpeta de trabajo `data/`.

A lo largo de este pipeline aplicaremos las siguientes fases lógicas:
1. **Capa Bronze (Ingesta Cruda):** Absorción directa de los archivos CSV respetando sus delimitadores nativos sin alterar sus estructuras.
2. **Capa Silver (Limpieza y Enriquecimiento):** Normalización de tipos de datos, corrección de anomalías en formatos numéricos regionales (comas decimales) y validación de consistencia.
3. **Capa Gold (Modelo Analítico de Negocio):** Generación de agregaciones estratégicas, cálculo de inventarios ideales y matrices de alertas operativas.

## 4.1. Capa Bronze: Ingesta de Datos Crudos

La capa Bronze representa el punto de entrada de la información al Lakehouse. Su propósito principal es persistir el histórico completo de los datos en formato **Delta Lake** para habilitar transacciones ACID y asegurar el linaje de datos (*Data Lineage*). 

**Estrategia de Ingesta:**
* Los catálogos maestros (`SITES.csv` e `ITEMS.csv`) se leen utilizando el punto y coma (`;`) como delimitador.
* El histórico de ventas (`BASE.csv`) y la foto de inventarios (`STOCK.csv`) se leen utilizando la coma (`,`) como delimitador.
* Todas las columnas se cargan inicialmente en formato de texto (`String`) para evitar truncamientos accidentales o pérdida de registros durante la fase de importación inicial.

```python
# Inicialización de bases de datos en el catálogo
spark.sql("CREATE DATABASE IF NOT EXISTS bigdataunaula.compras_bronze")
spark.sql("CREATE DATABASE IF NOT EXISTS bigdataunaula.compras_silver")
spark.sql("CREATE DATABASE IF NOT EXISTS bigdataunaula.compras_gold")

# Ruta raíz unificada donde se ubican los archivos cargados
ruta_raiz = "/Workspace/Users/kevinmateojimenez@gmail.com/TrabajoUNAULA_BIGDATA/Data/"

print("--> Iniciando ingesta de archivos hacia la Capa Bronze...")

# 1. Ingesta de SITES (Delimitador ';')
df_sites_raw = spark.read.option("header", "true").option("delimiter", ";").csv(f"{ruta_raiz}SITES.csv")
df_sites_raw.write.format("delta").mode("overwrite").saveAsTable("bigdataunaula.compras_bronze.tbl_sites")

# 2. Ingesta de ITEMS (Delimitador ';')
df_items_raw = spark.read.option("header", "true").option("delimiter", ";").csv(f"{ruta_raiz}ITEMS.csv")
df_items_raw.write.format("delta").mode("overwrite").saveAsTable("bigdataunaula.compras_bronze.tbl_items")

# 3. Ingesta de BASE (Delimitador ',')
df_base_raw = spark.read.option("header", "true").option("delimiter", ",").csv(f"{ruta_raiz}BASE.csv")
df_base_raw.write.format("delta").mode("overwrite").saveAsTable("bigdataunaula.compras_bronze.tbl_base")

# 4. Ingesta de STOCK (Delimitador ',')
df_stock_raw = spark.read.option("header", "true").option("delimiter", ",").csv(f"{ruta_raiz}STOCK.csv")
df_stock_raw.write.format("delta").mode("overwrite").saveAsTable("bigdataunaula.compras_bronze.tbl_stock")

print("--> ¡Éxito! Tablas creadas en el esquema 'bigdataunaula.compras_bronze'.")
```

## 4.2. Capa Silver: Transformación y Enriquecimiento de Datos

Una vez que los datos crudos están seguros en la capa Bronze, pasamos a la **Capa Silver**. El objetivo aquí es limpiar, tipificar y estructurar la información para que sea completamente confiable y analítica.

**Desafíos de Calidad de Datos Identificados y Resueltos:**
1. **Formato Decimal Regional:** Las columnas numéricas en `tbl_base` (`VentasSemestrales`, `Salidas`) y `tbl_stock` (`VentasSemestrales`, `Stock`) registran los decimales con comas (ej. `4232,3400`). Utilizaremos expresiones regulares (`regexp_replace`) para sustituir las comas por puntos y transformarlas formalmente a tipos numéricos de precisión flotante (`DoubleType`).
2. **Tipificación Estricta:** Las fechas en formato de texto se transformarán a `DateType` (`YYYY-MM-DD`), y las llaves primarias o foráneas (`ID`, `IdProducto`, `IdBodega`, `Ranking`) se convertirán a enteros (`IntegerType`).
3. **Normalización del Modelo:** Guardaremos estas tablas limpias bajo la nomenclatura estándar de almacenamiento empresarial: `dim_` para tablas maestras (dimensiones) y `fact_` para las tablas transaccionales (hechos).

```python
from pyspark.sql.functions import col, regexp_replace, trim

print("--> Iniciando transformaciones para la Capa Silver...")

# 1. Transformación de SITES -> dim_sites
df_sites_bronze = spark.table("bigdataunaula.compras_bronze.tbl_sites")
dim_sites = df_sites_bronze.select(
    col("ID").cast("integer").alias("id_tienda"),
    trim(col("NombreTienda")).alias("nombre_tienda")
)
dim_sites.write.format("delta").mode("overwrite").saveAsTable("bigdataunaula.compras_silver.dim_sites")


# 2. Transformación de ITEMS -> dim_items
df_items_bronze = spark.table("bigdataunaula.compras_bronze.tbl_items")
dim_items = df_items_bronze.select(
    col("ID").cast("integer").alias("id_producto"),
    trim(col("Nombre")).alias("nombre_producto"),
    trim(col("Proveedor")).alias("proveedor"),
    col("LeadTime").cast("integer").alias("lead_time_dias"),
    col("Precio").cast("double").alias("precio_unitario")
)
dim_items.write.format("delta").mode("overwrite").saveAsTable("bigdataunaula.compras_silver.dim_items")


# 3. Transformación de BASE -> fact_ventas
df_base_bronze = spark.table("bigdataunaula.compras_bronze.tbl_base")
fact_ventas = df_base_bronze.select(
    col("Ranking").cast("integer").alias("ranking_ventas"),
    col("IdProducto").cast("integer").alias("id_producto"),
    # Corrección de comas decimales regionales
    regexp_replace(col("VentasSemestrales"), ",", ".").cast("double").alias("ventas_semestrales_valor"),
    col("fecha").cast("date").alias("fecha_transaccion"),
    col("IdBodega").cast("integer").alias("id_tienda"),
    regexp_replace(col("Salidas"), ",", ".").cast("double").alias("cantidad_salidas")
)
fact_ventas.write.format("delta").mode("overwrite").saveAsTable("bigdataunaula.compras_silver.fact_ventas")


# 4. Transformación de STOCK -> fact_stock
df_stock_bronze = spark.table("bigdataunaula.compras_bronze.tbl_stock")
fact_stock = df_stock_bronze.select(
    col("Ranking").cast("integer").alias("ranking_inventario"),
    col("IdProducto").cast("integer").alias("id_producto"),
    regexp_replace(col("VentasSemestrales"), ",", ".").cast("double").alias("ventas_semestrales_valor"),
    regexp_replace(col("Stock"), ",", ".").cast("double").alias("stock_disponible")
)
fact_stock.write.format("delta").mode("overwrite").saveAsTable("bigdataunaula.compras_silver.fact_stock")

print("--> ¡Éxito! Capa Silver consolidada bajo el prefijo 'dim_' y 'fact_'.")
```

## 4.3. Capa Gold: Modelado de Indicadores y Alertas de Negocio

La capa Gold contiene datos agregados y optimizados para el consumo directo de los tableros de control (Dashboards) y los tomadores de decisiones de la compañía. En esta etapa, transformamos las tablas limpias de la capa Silver en respuestas a las 4 necesidades estratégicas del negocio:

1. **Clasificación de Inventario Ideal (Mínimos y Máximos):** Calcularemos la venta diaria promedio de cada producto (`ventas_semestrales_valor / 180 días`). Usando el `lead_time_dias` del proveedor, estableceremos el **Punto de Pedido** óptimo. A partir de allí, definiremos el *Inventario Mínimo Ideal* (como stock de seguridad) y el *Inventario Máximo Ideal* (para evitar sobrecostos de almacenamiento).
2. **Alertas de Criticidad (Stockouts y Excesos):** Todo producto con stock en 0 se marcará como `STOCKOUT`. Si un producto tiene stock por debajo del mínimo se catalogará como `Alta Necesidad`, si supera el máximo como `Exceso de Inventario`, y si está en los rangos correctos como `Inventario Saludable`.
3. **Priorización de Desabastecimiento:** Para los `STOCKOUTS`, cruzaremos el ranking de ventas; si el producto pertenece al grupo de alta demanda, se le asignará `Prioridad Crítica` para acelerar su compra.
4. **Ranking de Proveedores:** Evaluaremos el desempeño comercial consolidando las ventas totales (unidades vendidas por precio unitario) y salidas históricas asignadas a cada proveedor listado en el maestro de ítems.

```python
from pyspark.sql.functions import col, when, lit, round, sum

print("--> Generando indicadores estratégicos para la Capa Gold...")

# Cargar tablas de la capa Silver necesarias
dim_items = spark.table("bigdataunaula.compras_silver.dim_items")
fact_stock = spark.table("bigdataunaula.compras_silver.fact_stock")
fact_ventas = spark.table("bigdataunaula.compras_silver.fact_ventas")

# =========================================================================
# REGLA 1, 2 y 3: Matriz de Alertas de Abastecimiento (Mínimos, Máximos y Stockouts)
# =========================================================================

# Cruzamos la foto de stock actual con el maestro de productos para obtener LeadTime y Precio
df_analisis_stock = fact_stock.join(dim_items, "id_producto", "inner")

# Calculamos parámetros logísticos clave basados en los patrones de venta e inventario actual
df_alertas = df_analisis_stock.withColumn(
    "venta_diaria_promedio", round(col("ventas_semestrales_valor") / 180, 4)
).withColumn(
    "punto_de_pedido", round(col("venta_diaria_promedio") * col("lead_time_dias"), 4)
).withColumn(
    "inventario_minimo_ideal", round(col("punto_de_pedido") * 0.5, 4) # 50% del punto de pedido como stock de seguridad
).withColumn(
    "inventario_maximo_ideal", round(col("punto_de_pedido") + (col("venta_diaria_promedio") * 30), 4) # Punto de pedido + 30 días de venta
)

# Aplicamos la lógica de clasificación de alertas y prioridades solicitadas
df_alertas_final = df_alertas.withColumn(
    "estado_inventario",
    when(col("stock_disponible") == 0, lit("STOCKOUT"))
    .when(col("stock_disponible") < col("inventario_minimo_ideal"), lit("Alta Necesidad"))
    .when(col("stock_disponible") > col("inventario_maximo_ideal"), lit("Exceso de Inventario"))
    .otherwise(lit("Inventario Saludable"))
).withColumn(
    "prioridad_compra",
    when((col("stock_disponible") == 0) & (col("ranking_inventario") <= 500), lit("1. CRÍTICA (Alta Demanda)"))
    .when(col("stock_disponible") == 0, lit("2. ALTA"))
    .when(col("stock_disponible") < col("inventario_minimo_ideal"), lit("3. MEDIA"))
    .otherwise(lit("4. BAJA / REGULAR"))
)

# Guardar resultado en la capa Gold
df_alertas_final.write.format("delta").mode("overwrite").saveAsTable("bigdataunaula.compras_gold.agg_alertas_abastecimiento")


# =========================================================================
# REGLA 4: Ranking de Proveedores Comercial
# =========================================================================

# Cruzamos el histórico de ventas con ítems para saber qué proveedor vendió más dinero
df_ventas_proveedor = fact_ventas.join(dim_items, "id_producto", "inner")

rep_ranking_proveedores = df_ventas_proveedor.groupBy("proveedor").agg(
    round(sum(col("cantidad_salidas")), 2).alias("total_unidades_salidas"),
    round(sum(col("cantidad_salidas") * col("precio_unitario")), 2).alias("total_ingresos_generados")
).orderBy(col("total_ingresos_generados").desc())

# Guardar reporte en la capa Gold
rep_ranking_proveedores.write.format("delta").mode("overwrite").saveAsTable("bigdataunaula.compras_gold.rep_ranking_proveedores")

print("--> ¡Éxito! Capa Gold finalizada. Tablas 'agg_alertas_abastecimiento' y 'rep_ranking_proveedores' creadas.")
```

```python
# Verificación rápida de las alertas prioritarias generadas
print("Muestra de Productos en STOCKOUT o con Necesidad de Compra:")
display(spark.table("compras_gold.agg_alertas_abastecimiento")
        .select("id_producto", "nombre_producto", "stock_disponible", "estado_inventario", "prioridad_compra")
        .orderBy("prioridad_compra")
        .limit(5))
```

## 5.3. Capa Gold: Indicadores y Métricas de Negocio
Visualización de las tablas agregadas y optimizadas para análisis de negocio, incluyendo alertas de abastecimiento y KPIs estratégicos.

```python
from pyspark.sql.functions import col

print("=" * 80)
print("CAPA GOLD - INDICADORES Y MÉTRICAS DE NEGOCIO")
print("=" * 80)

# agg_alertas_abastecimiento
print("\n🚨 Agregado: bigdataunaula.compras_gold.agg_alertas_abastecimiento")
print(f"Total registros: {spark.table('bigdataunaula.compras_gold.agg_alertas_abastecimiento').count()}")
print("\nTop 10 productos con mayor prioridad de compra:")
display(spark.table("bigdataunaula.compras_gold.agg_alertas_abastecimiento")
        .select("id_producto", "nombre_producto", "stock_disponible", "venta_diaria_promedio", 
                "inventario_minimo_ideal", "inventario_maximo_ideal", "estado_inventario", "prioridad_compra")
        .orderBy(col("prioridad_compra"))
        .limit(10))

print("\n📊 Resumen de Productos por Estado de Inventario:")
display(spark.table("bigdataunaula.compras_gold.agg_alertas_abastecimiento")
        .groupBy("estado_inventario")
        .count()
        .orderBy("estado_inventario"))

print("\n⚠️ Resumen por Prioridad de Compra:")
display(spark.table("bigdataunaula.compras_gold.agg_alertas_abastecimiento")
        .groupBy("prioridad_compra")
        .count()
        .orderBy("prioridad_compra"))
```
