# 1. CASO DE NEGOCIO 

## 1.1. Descripción del Problema 
En el entorno actual del comercio minorista (Retail), la gestión ineficiente de la cadena de suministro y la falta de visibilidad en tiempo real de los inventarios representan desafíos críticos que impactan directamente la rentabilidad y la satisfacción del cliente. Actualmente, la organización enfrenta dificultades para consolidar y procesar de manera ágil grandes volúmenes de datos distribuidos en múltiples fuentes de información dispersas: catálogos de productos (`ITEMS`), datos maestros de tiendas (`SITES`), registros históricos de transacciones (`BASE`) y niveles actuales de existencias (`STOCK`).

Esta desconexión y la falta de analítica avanzada provocan dos problemas operacionales severos:
1. **Quiebres de Stock (Stockouts):** Desabastecimiento de productos de alta demanda sin una priorización clara de reposición, lo que se traduce en ventas perdidas y migración de clientes hacia la competencia.
2. **Exceso de Inventario (Overstock):** Acumulación ineficiente de capital de trabajo y costos de almacenamiento en productos de baja rotación, debido a la ausencia de umbrales máximos y mínimos dinámicos por cada binomio artículo-tienda (Item-Site).

La falta de un sistema predictivo y centralizado impide reaccionar proactivamente ante los patrones de venta de los consumidores y evaluar el desempeño de los proveedores de manera automatizada.

## 1.2. Objetivos del Proyecto 
El objetivo principal de este proyecto es implementar una solución integral de Big Data en **Databricks**, utilizando una **Arquitectura Medallion** (Bronze, Silver y Gold) para centralizar, transformar y explotar los datos de inventario y ventas, transformándolos en decisiones estratégicas de compra.

[cite_start]Con el análisis de datos masivos y el despliegue de un pipeline automatizado, se espera lograr el siguiente impacto[cite: 6]:
* **Optimización de Disponibilidad:** Identificar y priorizar de forma automática los *stockouts* de productos con alta demanda para mitigar el impacto financiero en el negocio.
* **Abastecimiento Proactivo e Inteligente:** Desarrollar un modelo analítico capaz de proyectar alertas de compra anticipadas basándose en los patrones históricos de venta (`BASE`), evitando llegar al quiebre de stock.
* **Control Eficiente de Inventarios:** Clasificar dinámicamente cada relación *Item-Site* mediante políticas de inventario ideal (Mínimos y Máximos) para detectar excesos de stock y necesidades críticas de reabastecimiento.
* **Desarrollo de un Dashboard de Control Centralizado:** Proveer un tablero de control interactivo para los tomadores de decisiones que unifique las alertas operativas y un *Ranking de Proveedores* automatizado para negociaciones más estratégicas.
* **Reducción de Costos Operativos:** Minimizar las horas-hombre dedicadas a la planeación manual de compras y optimizar los costos logísticos derivados del almacenamiento de excesos.

# 2. RELACIÓN BENEFICIO/COSTE

## 2.1. Análisis Económico (Ahorro Operativo)
Actualmente, el cálculo de necesidades de reabastecimiento, la consolidación de los archivos (`ITEMS`, `SITES`, `BASE`, `STOCK`) y la estimación de máximos y mínimos consumen una gran cantidad de tiempo operativo. Se estima que un equipo de 3 analistas de planeación de compras dedica alrededor de 60 horas semanales (240 horas al mes) exclusivamente a la limpieza, cruce de datos en hojas de cálculo y revisión manual de quiebres de stock.

Al automatizar este proceso mediante un pipeline de datos en **Databricks**, el procesamiento tomará escasos minutos de forma diaria y desatendida. Esto representa:
* **Ahorro en horas-hombre:** Reducción del 85% del tiempo operativo rutinario (aprox. 204 horas mensuales ahorradas).
* **Reasignación de talento:** Los especialistas de cadena de suministro dejarán de operar como "procesadores de datos" para enfocarse en la toma de decisiones estratégicas, negociación con proveedores y gestión de excepciones.

## 2.2. Retorno de Inversión (ROI)
El proyecto requiere una inversión inicial asociada a la configuración del entorno en la nube, el almacenamiento y el clúster de cómputo (Databricks), así como las horas de desarrollo e implementación técnica. Sin embargo, los beneficios tangibles compensan con creces estos costos:

* **Costo de Implementación y Operación (Proyectado - Año 1):** Inversión en arquitectura Cloud (Storage + Cómputo) y desarrollo, estimada en USD $15,000 (o su equivalente local).
* **Ahorros Proyectados (Año 1):** 
    1. Reducción de costos por sobreinventario (Overstock) al ajustar los máximos permitidos, liberando capital de trabajo estancado.
    2. Mitigación de los costos logísticos por almacenamiento innecesario.
* **Proyección de ROI:** Se estima que el proyecto alcance su punto de equilibrio en el mes 4 de operación, logrando un ROI superior al 150% al finalizar el primer año, impulsado por la eficiencia del flujo de caja.

## 2.3. Mayores Ingresos (Incremento en Ventas)
El impacto más significativo de este proyecto no solo radica en ahorrar, sino en **vender más**. 

Al implementar alertas automatizadas de *stockouts* para productos de alta demanda, la compañía evitará la pérdida de oportunidades de venta.
* **Ingresos vs Costos de Implementación:** Según datos del sector retail, las ventas perdidas por agotados pueden representar entre el 4% y el 8% de las ventas totales. Al contar con un sistema predictivo que genere alertas de pedidos antes del quiebre basándose en patrones (`BASE`), se proyecta recuperar al menos un 50% de esas ventas perdidas.
* **Conclusión Financiera:** El ingreso adicional neto generado por tener "el producto correcto en el momento adecuado" es exponencialmente mayor al costo de procesamiento diario del clúster de Databricks. La arquitectura propuesta se paga a sí misma exclusivamente con el incremento del volumen de ventas.

## 3.2. Explicación Técnica de los Componentes de Nube
Para dar soporte a la solución de optimización de compras y gestión de inventarios, se propone una arquitectura de **Lakehouse** unificada en la nube. Aunque el diseño empresarial ideal contempla una conexión directa y automatizada, el entorno actual ha sido adaptado mediante una zona de aterrizaje temporal debido a restricciones de acceso de red y políticas de seguridad corporativas:

1. **Origen de Datos Transaccional (SQL Server):** Representa el sistema de origen central (ERP o WMS de la compañía) donde residen originalmente las tablas operativas del negocio. Debido a las restricciones técnicas actuales, los datos son extraídos de forma asíncrona hacia archivos planos con formato `.csv`.
2. **Capa de Almacenamiento (Cloud Object Storage - Landing Zone):** Repositorio de objetos en la nube que actúa como la zona de aterrizaje para las extracciones de SQL Server (`ITEMS.csv`, `SITES.csv`, `BASE.csv` y `STOCK.csv`) antes de que el motor analítico inicie su lectura.
3. **Entorno de Cómputo Analítico (Databricks Workspace):** Entorno unificado basado en **Apache Spark** que se encarga de realizar la lectura distribuida de los archivos, la limpieza pesada, los cruzados relacionales (`JOINs`) y el procesamiento de las reglas de negocio a gran escala.
4. **Formato de Almacenamiento Lógico (Delta Lake):** Tecnología de almacenamiento de código abierto que corre sobre el Object Storage e introduce características de bases de datos relacionales tradicionales, tales como transacciones ACID, control de versiones (Time Travel) y validación de esquemas.
5. **Capa de Visualización y Consumo (Databricks SQL & Dashboards):** Herramienta nativa integrada que consumirá directamente las tablas agregadas de la capa Gold para renderizar de manera interactiva las métricas del negocio y disparar las alertas operativas.

## 3.3. Flujo de Información (Estrategia Medallion)
El refinamiento y la maduración de los datos procedentes de SQL Server se estructuran en tres niveles lógicos dentro de Databricks:

* **Fase de Ingesta (Capa Bronze - Datos Crudos):** Los archivos CSV se importan respetando estrictamente sus formatos de origen (manejando los delimitadores por coma y punto y coma correspondientes) y se persisten como tablas Delta nativas. No se altera la información para mantener el linaje completo y permitir auditorías ante fallos.
* **Fase de Enriquecimiento (Capa Silver - Limpieza y Cruces):** En esta etapa se ejecuta la tipificación estricta de las columnas. Se corrigen las anomalías regionales de formato (conversión de texto con comas decimales a números numéricos de tipo `Double` o `Float`) y se resuelven las discrepancias de consistencia lógica entre el histórico de ventas (`BASE`), las existencias de inventario (`STOCK`), y los catálogos maestros (`ITEMS` y `SITES`).
* **Fase de Negocio (Capa Gold - Agregaciones y Alertas):** Las tablas enriquecidas se transforman en indicadores directamente accionables por el área de compras. Se calculan los niveles críticos por tienda-producto (Stockouts y Excesos), los umbrales de inventario ideal (mínimos y máximos) y se consolida el *Ranking de Proveedores*.

## 3.4. Especificaciones del Clúster de Cómputo (Databricks Compute)
Para garantizar que el procesamiento de las transacciones (especialmente los miles de registros del histórico en `BASE`) se realice de manera eficiente, se define la siguiente configuración de infraestructura:

* **Modo de Clúster:** *Single Node* para mitigar costos durante la fase de desarrollo y laboratorio académico, proyectando un escalado a *Multi-node con Autoscaling* (2 a 8 Workers) para el despliegue final en producción.
* **Databricks Runtime (DBR):** Versión 14.3 LTS (Long Term Support) con soporte para Apache Spark 3.5.0 y Scala 2.12.
* **Motor de Rendimiento:** Activación obligatoria de **Photon Engine**, el motor de ejecución vectorizado de Databricks escrito en C++, optimizado para acelerar de forma drástica las consultas, filtros y agrupaciones analíticas complejas.
* **Política de Terminación Automática:** Configurada a 20 minutos de inactividad para optimizar el consumo de créditos de nube cuando el notebook o tablero no esté en uso.
